```python
import pandas as pd
from neuralprophet import NeuralProphet
from sklearn import svm 
from matplotlib import pyplot as plt
```

**Loading the dataset**

```python
# read data
data = pd.read_csv("seattle-weather.csv")
# to forcast multiple data, you must use them multiple times and each time you can forcast one target

# # date -> X , temp_max -> Y
# plt.plot(data ['date'], data ['temp_max'])
# plt.show()

# # date -> X , temp_min -> Y
# plt.plot(data ['date'], data ['temp_min'])
# plt.show()

date_temp_max = data[['date', 'temp_max']] 
date_temp_min = data[['date', 'temp_min']] 
# date_weather = data[['date', 'weather']] 
date_wind = data[['date', 'wind']] 
date_precipitation = data[['date', 'precipitation']] 

# Return object with labels on given axis omitted where alternately any or all of the data are missing
date_temp_max.dropna(inplace=True)
date_temp_max.columns = ['ds', 'y']      # you must name values like this

date_temp_min.dropna(inplace=True)
date_temp_min.columns = ['ds', 'y']      # you must name values like this

# date_weather.dropna(inplace=True)
# date_weather.columns = ['ds', 'y']      # you must name values like this

date_wind.dropna(inplace=True)
date_wind.columns = ['ds', 'y']      # you must name values like this

date_precipitation.dropna(inplace=True)
date_precipitation.columns = ['ds', 'y']      # you must name values like this
```

**Training the forecasting model**

```python
n1 = NeuralProphet()
n2 = NeuralProphet()
n3 = NeuralProphet()
n4 = NeuralProphet()
n5 = NeuralProphet()
model_max = n1.fit(date_temp_max, freq='D', epochs=900)
model_min = n2.fit(date_temp_min, freq='D', epochs=900)
# model_weather = n3.fit(date_weather, freq='D', epochs=900)
model_wind = n4.fit(date_wind, freq='D', epochs=900)
model_precipitation = n5.fit(date_precipitation, freq='D', epochs=900)
      # training our model for 900 epochs
      # freq='D' -> we're using a daily frequency
      # training our model for 900 epochs
```

**Forecasting into the future**

```python
future_temp_max = n1.make_future_dataframe(date_temp_max, periods=1500)
future_temp_min = n2.make_future_dataframe(date_temp_min, periods=1500)
# future_weather = n3.make_future_dataframe(date_weather, periods=1500)
future_wind = n4.make_future_dataframe(date_wind, periods=1500)
future_precipitation = n5.make_future_dataframe(date_precipitation, periods=1500)
# forcasting weather for 1500 dayes into the future

forecast_max = n1.predict(future_temp_max)  # to predict our future values
forecast_min = n2.predict(future_temp_min)
# forecast_weather = n3.predict(future_weather)
forecast_wind = n4.predict(future_wind)
forecast_precipitation = n5.predict(future_precipitation)

# forecast.tail()               # to list our five last rows

plot = n1.plot(forecast_max)
plot.show()

plot = n2.plot(forecast_min)
plot.show()

# plot = n3.plot(forecast_weather)
# plot.show()

plot = n4.plot(forecast_wind)
plot.show()

plot = n5.plot(forecast_precipitation)
plot.show()
```
